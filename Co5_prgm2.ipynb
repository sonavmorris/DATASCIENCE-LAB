```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
print(soup.prettify())# prettify() enable us to view how tags are nested in that document
```

```python
product=soup.find_all('div',class_="_4rR01T")
product
```

```python
mob_names=[]                   #list of products
for i in range(0,len(product)):
 mob_names.append(product[i].get_text())
```

```python
mob_names
```

```python
price=soup.find_all('div',class_='_30jeq3 _1_WHN1')
price
```

```python
pricelist=[]                   #list of products
for i in range(0,len(price)):
 pricelist.append(price[i].get_text())
pricelist
```

```python
rating=soup.find_all('div',class_='_3LWZlK')
rating
```

```python
ratinglist=[]                   #list of products
for i in range(0,len(rating)):
 ratinglist.append(rating[i].get_text())
ratinglist
```
